```python
# 1 D) Ã‰ possÃ­vel encontrar outliers atravÃ©s dos boxplots? Indique quais sÃ£o e remova-os caso julgar necessÃ¡rio
# Remover outliers Multiplicar dados por atÃ© 3x pelo Desvio PadrÃ£o;
# Fazer para Radius, Perimetro e Area 
import numpy as np


df_clean = dfm[['id', 'diagnosis','radius_mean', 'perimeter_mean', 'area_mean']]

df_clean = df_clean[ np.abs( df_clean.radius_mean - df_clean.radius_mean.mean()) <= (2.5 * df_clean.radius_mean.std() )]
df_clean = df_clean[ np.abs( df_clean.perimeter_mean - df_clean.perimeter_mean.mean()) <= (2.5 * df_clean.perimeter_mean.std() )]
df_clean = df_clean[ np.abs( df_clean.area_mean - df_clean.area_mean.mean()) <= (2.5 * df_clean.area_mean.std() )]

df_clean.head()
```

```python
plt.figure(figsize=[20,8])
plt.subplot(2,3,1)
plt.title('radius original')
plt.boxplot(dfm.radius_mean)

plt.subplot(2,3,2)
plt.title('perimeter original')
plt.boxplot(dfm.perimeter_mean)

plt.subplot(2,3,3)
plt.title('area original')
plt.boxplot(dfm.area_mean)

plt.subplot(2,3,4)
plt.title('radius clean')
plt.boxplot(df_clean.radius_mean)

plt.subplot(2,3,5)
plt.title('perimeter clean')
plt.boxplot(df_clean.perimeter_mean)

plt.subplot(2,3,6)
plt.title('area clean')
plt.boxplot(df_clean.area_mean)
```

```python
# Em relaÃ§Ã£o aos outliers a variavel Ã¡rea possui alguns bem distantes. F
# Foram removidos limitando os valores atÃ© 3x o desvio padrÃ£o, o mesmo critÃ©rio foi aplicar para raiuds e perimetro
```

```python
#apenas 3 colunas

x = df_clean[['area_mean', 'perimeter_mean', 'radius_mean' ]].values
y = df_clean[['diagnosis']].values
y = df_clean.diagnosis.values

plt.scatter(x[:,0], x[:,1], x[:,2], c = y)

```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
xtreino, xteste, ytreino, yteste = train_test_split(x, y, test_size = 0.4)

svc = SVC(probability=True, kernel='linear')
classif = svc.fit(xtreino, ytreino)
probs = classif.predict_proba(xteste)

data = np.c_[xteste, yteste, probs[:,0], probs[:,1]]
data_df = pd.DataFrame(data)
data_df.columns = ['area', 'perimeter', 'radius', 'diagnosis', 'prob_tipo_B' , 'prob_tipo_M']

print(' ------------  print do data_df ------------')
print()
print(data_df.tail(20))
print("-------------------------------------------------------------------")
print('acurÃ¡cia do modelo SVM: ' ,accuracy_score(classif.predict(xteste), yteste))
print("-------------------------------------------------------------------")
print()
```

```python
# AcurÃ¡cia aproximada de 90% usando o modelo SVM - Linear, porÃ©m com as variaveis na seguinte ordem:
# 1 area_mean , 2 - perimeter_mean, 3 - radius_Mean
```

```python
# Verificar acuÃ¡rica com modelo SVC(kernel= â€™rbfâ€™)

xtreino, xteste, ytreino, yteste = train_test_split(x, y, test_size = 0.4)

svc = SVC(probability=True, kernel='rbf')
classif = svc.fit(xtreino, ytreino)
probs = classif.predict_proba(xteste)

data = np.c_[xteste, yteste, probs[:,0], probs[:,1]]
data_df = pd.DataFrame(data)
data_df.columns = ['area', 'perimeter', 'radius', 'diagnosis', 'prob_tipo_B' , 'prob_tipo_M']

print(' ------------  print do data_df ------------')
print()
print(data_df.tail(20))
print("-------------------------------------------------------------------")
print('acurÃ¡cia do modelo SVM: ' ,accuracy_score(classif.predict(xteste), yteste))
print("-------------------------------------------------------------------")
print()
```

```python
def plot_compara_real_vs_predito(X, y, modelo, split_size = 0.3):
    xtreino, xteste, ytreino, yteste = train_test_split(x, y, test_size = 0.4)
    modelo.fit(xtreino,ytreino)
    preds = modelo.predict(xteste)
    
    plt.figure(figsize=[20,5])
    plt.subplot(1,2,1)
    plt.scatter(xteste[:,0], xteste[:,1], c = yteste, cmap = 'viridis', s= 140)
    plt.title("DistribuiÃ§Ã£o real dos dados de teste")
    
    plt.subplot(1,2,2)
    plt.scatter(xteste[:,0], xteste[:,1], c = preds, cmap = 'viridis', s=140)
    plt.title("PrediÃ§Ã£o dos dados de teste. AcurÃ¡cia: " + str(round(float(accuracy_score(preds, yteste)),3)))
    
    plt.show()
```
